```python
#splitting the dataset into numerical and categorical
df_num=df_pbm1.select_dtypes(['float64','int64'])
df_cat=df_pbm1.select_dtypes(['object'])
```

```python
#filling bulk null values with median
from sklearn.impute import SimpleImputer
imup=SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
df_num=pd.DataFrame(imup.fit_transform(df_num),columns=df_num.columns)
```

```python
df_pbm1=pd.concat([df_num,df_cat],axis=1,join='inner')
```
